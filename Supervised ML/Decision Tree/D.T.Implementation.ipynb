```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris =load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent and dependent features
x= df.iloc[:,:-1]
```

```python
y= iris.target
```

```python
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel= DecisionTreeClassifier()
```

## Post Pruning

```python
treemodel.fit(x_train, y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(treemodel,filled=True)
```

Now we will tune it as we have seen that it will be enough to apply depth upto level 2

```python
plt.figure(figsize=(12,8))
tree.plot_tree(treemodel,filled=True, max_depth=2)
```

```python
y_pred= treemodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(y_pred, y_test)*100)
print(classification_report(y_pred, y_test))
```

## Pre Pruning

```python
parameters= {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features':['auto', 'sqrt','log2']

}
```

```python
# apply grid search cv
from sklearn.model_selection import GridSearchCV
```

```python
treemodel2 = DecisionTreeClassifier()
cv = GridSearchCV(treemodel2, param_grid= parameters, cv= 5, scoring='accuracy')
```

```python
cv.fit(x_train, y_train)
```

```python
cv.best_estimator_
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
y_pred=cv.predict(x_test)
```

```python
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
#visualization
model = DecisionTreeClassifier(criterion= 'gini',
 max_depth= 4,
 max_features= 'log2',
 splitter= 'best')
```

```python
model.fit(x_train, y_train)
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(model,filled=True)
```

### best estimators: actually train the model too
### best params: just tell the best parameters.....you have to actually train the model


```python
## now using best estimators
best_model = cv.best_estimator_
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8))
plot_tree(best_model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Pre-Pruned Decision Tree for Iris Dataset (Best Model from Grid Search CV)")
plt.show()
```
