```python
import pandas as pd

# Load the CSV file, specifying the correct separator and encoding
messages = pd.read_csv('spam.csv', sep=',', header=0, encoding='latin-1')

# Display the first few rows of the DataFrame
print("Original DataFrame:")
print(messages.head())

# Select only the relevant columns (first two columns)
messages = messages.iloc[:, [0, 1]]  # Keep only the first two columns

# Rename the columns
messages.columns = ['label', 'message']

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(messages.head())
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

### Creating the Bag of Words Model

```python
#converting dependent features to zero and one

y= pd.get_dummies( messages['label'], dtype=int)
y= y.iloc[:, 0].values
```

```python
#train test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(corpus,y, test_size= 0.20)
```

```python
x_train
```

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
x_train = cv.fit_transform(x_train).toarray()
x_test = cv.transform(x_test).toarray()
```

### Creating the TF-IDF Model

```python
#train test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(corpus,y, test_size= 0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv= TfidfVectorizer(max_features=2500, ngram_range=(1,2))
x_train = tv.fit_transform(x_train).toarray()
x_test = tv.transform(x_test).toarray()
```

```python
tv
```

```python
spam_tfidf_model = MultinomialNB().fit(x_train, y_train)
```

```python
y_pred2 = spam_tfidf_model.predict(x_test)
```

```python
y_pred2
```

```python
accuracy_score( y_test, y_pred2)
```

```python
print(classification_report( y_test, y_pred2))
```
