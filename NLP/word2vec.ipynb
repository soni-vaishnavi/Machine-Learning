```python

import pandas as pd
import re
# Load the CSV file, specifying the correct separator and encoding
messages = pd.read_csv('spam.csv', sep=',', header=0, encoding='latin-1')

# Display the first few rows of the DataFrame
print("Original DataFrame:")
print(messages.head())

# Select only the relevant columns (first two columns)
messages = messages.iloc[:, [0, 1]]  # Keep only the first two columns

# Rename the columns
messages.columns = ['label', 'message']

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(messages.head())
```

```python
from nltk.stem import WordNetLemmatizer as wnl
```

```python
lemmatizer = wnl()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-z]',' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize, word_tokenize ## you can use simple processess instead of word_tokenize, it lowers, removes punctuations
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token= sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
## lets train with word2vec from scratch

model = gensim.models.Word2Vec(words)
```
